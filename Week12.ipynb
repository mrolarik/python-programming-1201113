```
import numpy as np

# ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å (kg) ‡πÅ‡∏•‡∏∞‡∏™‡πà‡∏ß‡∏ô‡∏™‡∏π‡∏á (m) ‡∏Ç‡∏≠‡∏á 5 ‡∏Ñ‡∏ô
weights = np.array([55, 70, 60, 80, 95])  # ‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å
heights = np.array([1.65, 1.75, 1.60, 1.80, 1.75])  # ‡∏™‡πà‡∏ß‡∏ô‡∏™‡∏π‡∏á

# ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì BMI
bmi = weights / (heights ** 2)

# ‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏™‡∏ñ‡∏≤‡∏ô‡∏∞‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å‡∏ï‡∏≤‡∏°‡∏Ñ‡πà‡∏≤ BMI
def categorize_bmi(bmi_values):
    categories = []
    for value in bmi_values:
        if value < 18.5:
            categories.append("‡∏ú‡∏≠‡∏°")
        elif 18.5 <= value < 24.9:
            categories.append("‡∏õ‡∏Å‡∏ï‡∏¥")
        elif 25 <= value < 29.9:
            categories.append("‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å‡πÄ‡∏Å‡∏¥‡∏ô")
        else:
            categories.append("‡∏≠‡πâ‡∏ß‡∏ô")
    return np.array(categories)

# ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏´‡∏°‡∏ß‡∏î‡∏´‡∏°‡∏π‡πà BMI
bmi_categories = categorize_bmi(bmi)

# ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå
for i in range(len(weights)):
    print(f"‡∏ö‡∏∏‡∏Ñ‡∏Ñ‡∏• {i+1}: ‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å {weights[i]} kg, ‡∏™‡∏π‡∏á {heights[i]} m, BMI: {bmi[i]:.2f}, ‡∏™‡∏ñ‡∏≤‡∏ô‡∏∞: {bmi_categories[i]}")

```
