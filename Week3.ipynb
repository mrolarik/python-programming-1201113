**If-Else statement**  
If = ถ้า  
Else = แล้ว  

If-Else = ถ้า-แล้ว  

```
if(condition):
    if_body()
else:
    else_body()
```

*   condition เช่น (a > b), (a == b)
*   ถ้าเงื่อนไขเป็นจริง (True) จะทำงานที่ if_body()
*   ถ้าเงื่อนไขเป็นเท็จ (False) จะทำงานที่ else_body()

เช่น  



```
a = 5
b = 6

if(a>b):
    print(a, "is larger than", b)
else:
    print(b, "is larger than", b)
```




ตัวอย่างโปรแกรม

````python
a = 5
b = 6

if(a>b):
  print(a, "is larger than", b)
else:
  print(b, "is larger than", a)
````

**Logic Gate**

AND gate

image.png

**AND gate**
จะมีอยู่เงื่อนไขเดียวคือ a และ b ต้องมีค่าเป็นจริง (True) ทั้งคู่  
  
*AND operation*  

````python
a = 1
b = 0

if(a and b):
  print("output is 1")
else:
  print("output is 0")
````

````python
a = 1
b = 0

if(a & b):
  print("output is 1")
else:
  print("output is 0")
````

OR gate

image.png

**OR Gate** จะมีค่าเป็นเท็จเมื่อทั้ง a และ b มีค่าเป็นเท็จ  
  
*OR operation*  

````python
a = 1
b = 0

if(a or b):
  print("output is 1")
else:
  print("output is 0")
````

````python
a = 1
b = 0

if(a | b):
  print("output is 1")
else:
  print("output is 0")
````
