**Syntax**

```
initializing variable
while(expression):
  statement 1
  statement 2
  increasing variable
```

ตัวอย่างโค้ดของ While-Loop

```
#initializing variable (กำหนดค่าเริ่มต้นให้กับตัวแปรที่จะใช้ในเงื่อนไขของ while-loop)
i = 0

#expression (เงื่อนไขที่จะให้ while-loop ทำงาน)
while(i<10):
  #statement
  print(i)

  #increasing variable (เพิ่มค่าให้กับตัวแปรที่ใช้ในเงื่อนไข)
  i = i+1
```


**กรณีที่เงื่อนไขเป็นจริง While-loop จะทำงาน**
* โดยเริ่มต้นกำหนดให้ `i=0` และเงื่อนไขคือ `i<10` ดังนั้น ในรอบแรก i มีค่าเท่ากับ 0 และ i มีค่าน้อยกว่า 10 (i<10) โปรแกรมจึงเข้าไปทำงานในลูป (While-loop)

````python
i=0
while(i<10):
  print(i)
  i = i+1
````

**กรณีที่กำหนดเงื่อนไขผิด** โปรแกรมจะไม่ทำงาน
* เช่น หากกำหนดให้ `i=0` และเงื่อนไขในการทำงานของลูปคือ `i>10` และในกรณีนี้ **i มีค่าน้อยกว่า 10** โปรแกรมจึงไม่เข้าไปทำงานในลูป (While-loop)
* จากตัวอย่างต่อไปนี้ โปรแกรมจึงไม่เข้าไปในลูปเพื่อ `print` ค่า `i`

````python
i=0
while(i>10):
  print(i)
  i = i+1
````

**ตัวอย่าง** การประยุกต์ใช้ While-loop เพื่อสร้างเมนูแสดงการทำงาน เช่น

```
Please select a menu option:
1. Addition (+)
2. Subtraction (-)
3. Multiplication (*)
4. Division (/)
5. Exit the program

I select:
```

จากตัวอย่าง
* หากเลือกหมายเลข `1` โปรแกรมจะแสดงข้อความ `You select number 1`
* หากเลือกหมายเลข `2` โปรแกรมจะแสดงข้อความ `You select number 2`
* หากเลือกหมายเลข `3` โปรแกรมจะแสดงข้อความ `You select number 3`
* หากเลือกหมายเลข `4` โปรแกรมจะแสดงข้อความ `You select number 4`
* หากเลือกหมายเลข `5` โปรแกรมจะแสดงข้อความ `You select number 5`

สามารถเขียนโปรแกรมได้ดังนี้

```
while(True):
  print("Please select a menu option:")
  print("1. Addition (+)")
  print("2. Subtraction (-)")
  print("3. Multiplication (*)")
  print("4. Division (/)")
  print("5. Exit the programprint")

  sel = int(input("I select: "))

  if(sel == 1):
    print("You select number 1")
  elif(sel == 2):
    print("You select number 2")
  elif(sel == 3):
    print("You select number 3")
  elif(sel == 4):
    print("You select number 4")
  elif(sel == 5):
    print("You select number 5")
    break
```

````python
while(True):
  print("Please select a menu option:")
  print("1. Addition (+)")
  print("2. Subtraction (-)")
  print("3. Multiplication (*)")
  print("4. Division (/)")
  print("5. Exit the programprint")

  sel = int(input("Select: "))

  if(sel == 1):
    print("Select number 1")
  elif(sel == 2):
    print("Select number 2")
  elif(sel == 3):
    print("Select number 3")
  elif(sel == 4):
    print("Select number 4")
  elif(sel == 5):
    print("Select number 5")
    break
````

**โจทย์** จากตัวอย่างโปรแกรมแสดงเมนูข้างต้น ให้เพิ่มเติมการเขียนโปรแกรมให้สมบูรณ์โดย
* หากผู้ใช้กดเลือกเมนู 1 ให้แสดงข้อความให้ป้อนตัวเลขตัวที่ 1 และ 2 และแสดงผลบวก (Addition) ของตัวเลขทั้ง 2 ตัว ดังนี้

```

1. Addition (+)
2. Subtraction (-)
3. Multiplication (*)
4. Division (/)
5. Exit the programprint
Select: 1
Enter the first number: 10
Enter the second number: 5
Result: 15
```

* หากผู้ใช้กดเลือกเมนู 2 ให้แสดงข้อความให้ป้อนตัวเลขตัวที่ 1 และ 2 และแสดงผลลบ (Subtraction) ของตัวเลขทั้ง 2 ตัว ดังนี้

```
Please select a menu option:
1. Addition (+)
2. Subtraction (-)
3. Multiplication (*)
4. Division (/)

```

* หากผู้ใช้กดเลือกเมนู 3 ให้แสดงข้อความให้ป้อนตัวเลขตัวที่ 1 และ 2 และแสดงผลคูณ (Multiplication) ของตัวเลขทั้ง 2 ตัว
* หากผู้ใช้กดเลือกเมนู 4 ให้แสดงข้อความให้ป้อนตัวเลขตัวที่ 1 และ 2 และแสดงผลหาร (Division) ของตัวเลขทั้ง 2 ตัว

* และหากผู้ใช้กดเลือกเมนู 5 ให้แสดงข้อความ `Thank you & Exit program!!` ดังนี้
