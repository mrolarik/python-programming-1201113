**List**
* ใช้สำหรับเก็บข้อมูล (Store item) ได้มากกว่า 1 ค่า โดยจัดเก็บไว้ในตัวแปรเพียง 1 ตัว
* ตัวอย่างการสร้างตัวแปรประเภท List (ลิสต์) เช่น  

*ตัวอย่าง 1*
```
my_list = [3, 4, 6, 10, 8]
```
* จากตัวอย่างข้างต้น สร้างตัวแปรชื่อ `my_list` โดยตัวแปรนี้เก็บข้อมูลทั้งหมด 5 ค่า คือ `3, 4, 6, 10, 8`
* ข้อมูลทั้ง 5 ค่าเป็นข้อมูลประเภทตัวเลขจำนวนเต็ม `integer (int)`

*ตัวอย่าง 2*
```
name = ["Olarik", "Kaveepoj", "Jackson"]
```
* จากตัวอย่างข้างต้น สร้างตัวแปรชื่อ `name` โดยตัวแปรนี้เก็บข้อมูลประเภทตัวอักษร `string (str)`
* ข้อมูลที่เก็บในตัวแปร `name` มีจำนวน 3 ข้อมูล ประกอบด้วย `"Olarik", "Kaveepoj", "Jackson"`

*ตัวอย่าง 3*
```
mix_list = [3, 4, "Olarik", [10.5, 6.2, 4.3]]
```
* จากตัวอย่างข้างต้น สร้างตัวแปรชื่อ `mix_list` โดยตัวแปรนี้เก็บข้อมูลมากกว่า 1 ประเภท ได้แก่ตัวเลข ตัวอักษร และลิสต์
* ข้อมูลที่เก็บในตัวแปร `mix_list` มีจำนวน 4 ข้อมูล

**Index (อินเด็กซ์)**
* การเข้าถึงข้อมูลในลิสต์ (Access List Items) จะต้องใช้อินเด็กซ์ (Index) ในการเข้าถึง โดย Index จะเป็นได้ทั้ง Positive Index และ Negative Index

ตัวอย่างการเข้าถึงลิสต์แสดงดังต่อไปนี้

image.png

จากตัวอย่างข้างต้น สามารถเขียนโปรแกรมได้ดังนี้

````python
My_List = [3, 4, 6, 10, 8]

print(My_List[0])
````

* จากคำสั่ง `My_List[0]` หมายถึงเข้าถึงข้อมูลในตัวแปร `My_List` โดยเข้าถึง Index ที่ 0
* โดยการเข้าถึง index จะต้องอยู่ในเครื่องหมาย `[]`
* ดังนั้น `My_List[0]` จึงหมายถึงเข้าถึงข้อมูลใน `My_List` ลำดับที่ 0 ซึ่งจะมีค่าคือ `3`

````python
My_List = [3, 4, 6, 10, 8]

print(My_List[-1])
````

* จากคำสั่ง My_List[-1] หมายถึงเข้าถึงข้อมูลในตัวแปร My_List * โดยเข้าถึง Index ที่ -1 (ลำดับสุดท้ายของ List)
* ดังนั้น My_List[-1] จึงหมายถึงเข้าถึงข้อมูลใน My_List ลำดับสุดท้าย ซึ่งจะมีค่าคือ `8`

ตรวจสอบประเภทของข้อมูลด้วย Built-in function `type()`

````python
print(type(My_List))
````

**เขียนโปรแกรม** ด้วย For-Loop เพื่อแสดงข้อมูลในลิสต์

````python
names = ["Olarik", "Kaveepoj", "Jackson"]

for i in range(len(names)):
  print(names[i])
````

````python
names = ["Olarik", "Kaveepoj", "Jackson"]

for name in names:
  print(name)
````

**เขียนโปรแกรม** ด้วย While-Loop เพื่อแสดงข้อมูลในลิสต์

````python
names = ["Olarik", "Kaveepoj", "Jackson"]

i = 0
while(i<len(names)):
  print(names[i])
  i = i+1
````

**สำหรับ For-Loop** - กรณีที่มี List จำนวนมากกว่า 1 ลิสต์ และต้องการแสดงข้อมูลสามารถใช้ built-in function ชื่อ `zip()` เข้ามาช่วย เช่น

````python
names = ["Olarik", "Kaveepoj", "Jackson"]
surnames = ["Surinta", "Bunluewong", "Wang"]

for name, surname in zip(names, surnames):
  print(name, surname)
````

**สำหรับ While-Loop** สามารถเขียนโปรแกรมเพื่อแสดงข้อมูลในลิสต์มากกว่า 1 ลิสต์ได้ ดังนี้

````python
names = ["Olarik", "Kaveepoj", "Jackson"]
surnames = ["Surinta", "Bunluewong", "Wang"]

i = 0
while(i<len(names)):
  print(names[i], surnames[i])
  i += 1
````
