# ‚úÖ ‡πÅ‡∏õ‡∏•‡∏á‡∏™‡∏µ‡∏£‡∏π‡∏õ‡∏†‡∏≤‡∏û (‡πÄ‡∏ä‡πà‡∏ô `RGB ‚Üí Grayscale`)  

````python
import matplotlib.pyplot as plt
from skimage import io, color

#‡πÇ‡∏´‡∏•‡∏î (load) ‡∏£‡∏π‡∏õ‡∏†‡∏≤‡∏û
image_path = "https://cdn.outsideonline.com/wp-content/uploads/2023/03/Funny_Dog_H.jpg"
image = io.imread(image_path)

# ‡πÅ‡∏õ‡∏•‡∏á‡πÄ‡∏õ‡πá‡∏ô‡∏†‡∏≤‡∏û‡∏™‡∏µ‡πÄ‡∏ó‡∏≤
grayscale = color.rgb2gray(image)

#‡πÅ‡∏™‡∏î‡∏á‡∏£‡∏π‡∏õ‡∏†‡∏≤‡∏û (show)
plt.imshow(grayscale, cmap="gray")
plt.axis("off")
plt.title("Grayscale Image")
plt.show()
````

# ‚úÖ ‡∏õ‡∏£‡∏±‡∏ö‡∏Ç‡∏ô‡∏≤‡∏î‡∏£‡∏π‡∏õ‡∏†‡∏≤‡∏û

````python
import matplotlib.pyplot as plt
from skimage import io, transform

#‡πÇ‡∏´‡∏•‡∏î (load) ‡∏£‡∏π‡∏õ‡∏†‡∏≤‡∏û
image_path = "https://cdn.outsideonline.com/wp-content/uploads/2023/03/Funny_Dog_H.jpg"
image = io.imread(image_path)

# ‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô‡∏Ç‡∏ô‡∏≤‡∏î‡∏£‡∏π‡∏õ‡∏†‡∏≤‡∏û
width, height = 200, 200
resized = transform.resize(image, (width, height))

plt.imshow(resized)
plt.axis("on")
plt.title(f"Resized Image ({width}x{height})")
plt.show()
````

````python
import matplotlib.pyplot as plt
from skimage import io, transform

image_path = "https://cdn.outsideonline.com/wp-content/uploads/2023/03/Funny_Dog_H.jpg"
image = io.imread(image_path)

original_height, original_width = image.shape[:2]

# ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏π‡∏á‡πÉ‡∏´‡∏°‡πà ‡πÇ‡∏î‡∏¢‡∏Ñ‡∏á‡∏≠‡∏±‡∏ï‡∏£‡∏≤‡∏™‡πà‡∏ß‡∏ô‡πÄ‡∏î‡∏¥‡∏°
new_width = 200
ratio = new_width / original_width
new_height = int(original_height * ratio)

# ‡∏õ‡∏£‡∏±‡∏ö‡∏Ç‡∏ô‡∏≤‡∏î‡∏†‡∏≤‡∏û
resized = transform.resize(image, (new_height, new_width), anti_aliasing=True)

plt.imshow(resized)
plt.axis("on")
plt.title(f"Resized Image ({width}x{height})")
plt.show()
````

# ‚úÖ ‡∏´‡∏°‡∏∏‡∏ô‡∏£‡∏π‡∏õ‡∏†‡∏≤‡∏û

````python
import matplotlib.pyplot as plt
from skimage import io, transform

image_path = "https://cdn.outsideonline.com/wp-content/uploads/2023/03/Funny_Dog_H.jpg"
image = io.imread(image_path)

angle = 20
rotated = transform.rotate(image, angle)
plt.imshow(rotated)
plt.axis("off")
plt.title(f"Rotated Image ({angle} degrees)")
plt.show()
````

# ‚úÖ ‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å‡∏£‡∏π‡∏õ‡∏†‡∏≤‡∏û

````python
import matplotlib.pyplot as plt
from skimage import io, transform, img_as_ubyte

image_path = "https://cdn.outsideonline.com/wp-content/uploads/2023/03/Funny_Dog_H.jpg"
image = io.imread(image_path)

angle = 20
rotated = transform.rotate(image, angle)

# Convert the rotated image to uint8 before saving
rotated = img_as_ubyte(rotated)

filename = "new_image.jpg"
io.imsave(filename, rotated)
print(f"‚úÖ ‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å‡∏£‡∏π‡∏õ‡∏†‡∏≤‡∏û‡πÄ‡∏õ‡πá‡∏ô‡πÑ‡∏ü‡∏•‡πå: {filename}")
````

# ‚úÖ ‡∏Å‡∏≤‡∏£‡πÅ‡∏õ‡∏•‡∏á‡∏†‡∏≤‡∏û‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô‡∏£‡∏π‡∏õ‡∏†‡∏≤‡∏û‡πÅ‡∏ö‡∏ö Edge Image

````python
import matplotlib.pyplot as plt
from skimage import io, transform, filters, feature

image_path = "https://cdn.outsideonline.com/wp-content/uploads/2023/03/Funny_Dog_H.jpg"
image = io.imread(image_path)

grayscale = color.rgb2gray(image)

# ‡∏ï‡∏£‡∏ß‡∏à‡∏à‡∏±‡∏ö‡∏Ç‡∏≠‡∏ö‡∏à‡∏≤‡∏Å‡∏†‡∏≤‡∏û
sigma=1.0
edges = feature.canny(grayscale, sigma=sigma)

plt.imshow(edges, cmap="gray")
plt.axis("off")
plt.title(f"Edge Image (Sigma={sigma})")
plt.show()
````
