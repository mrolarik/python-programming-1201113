# **If-Else-If statement**

## **Syntax**

```
if(condition_1):
  Body_of_IF
elif(condition_2):
  Body_of_ELIF
else:
  Body_of_else
```



## **flowchart**

image.png

## Sample code

image.png

จากตัวอย่าง flowchart ข้างต้น สามารถเขียนโปรแกรม `python` ได้ดังต่อไปนี้

* หมายเหตุ - flowchart อาจวาดไม่ถูกต้อง

````python
grade = 'null'
score = 85

if(score >= 90):    #หาก score >= 90 จะเข้าเงื่อนไขเป็นจริง (True) ดังนั้น grade = 'A'
  grade = 'A'       #เงื่อนไขเป็นจริงของ score >= 90
elif(score >= 80):  #หาก score >= 80 จะเข้าเงื่อนไขเป็นจริง (True) ดังนั้น grade = 'B'
  grade = 'B'       #เงื่อนไขเป็นจริงของ score >=80
elif(score >= 70):
  grade = 'C'
elif(score >= 60):
  grade = 'D'
else:
  grade = 'F'

print("Your grade is", grade)
````

**Nested-if**

image.png

````python
age = 21

if(age >= 11):
  print("Your are eligible to see the football match")
  if((age <=20) or (age >= 60)):      #Nested-if หมายถึง ถ้าอายุน้อยกว่าหรือเท่ากับ 20 หรือ อายุมากกว่าหรือเท่ากับ 60
    print("Ticket price is $12")      #ถ้าเป็นจริง จ่ายเงินค่าตั๋ว $12
  else:
    print("Ticket price is $20")      #ถ้าเป็นเท็จ จ่ายค่าตั๋ว $20 (หมายถึงอายุอยู่ในช่วง 21 ถึง 59)
else:
  print("You are not eligible to buy a ticket")
````
